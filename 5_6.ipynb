```python
# Необходимо написать функцию, в которой в бесконечном цикле генерируется 2 случайных натуральных числа (от 1 до 100)
# и в рамках математической игры запрашивается их сумма у пользователя. Если ответ пользователя неверный,
# то ему назначается штрафная пауза - он отстраняется от игры на какое-то количество секунд.
# Пусть изначально штрафная пауза составит 3 секунды, но с каждым неправильным ответом она будет увеличиваться на 1 секунду.
import random
import time
def fynkzia():
    shtraf = 3

    while True:
        a = random.randint(1, 100)
        b = random.randint(1, 100)

        print(f"\nНовый пример: {a} + {b}?")
        otvet = int(input())

        if otvet == a + b:
            print("Поздравляю! Ваш ответ правильный")
            break
        else:
            print(f"Неверно! Штрафная пауза: {shtraf} секунды")
            time.sleep(shtraf)
            shtraf = shtraf + 1
fynkzia()
```
