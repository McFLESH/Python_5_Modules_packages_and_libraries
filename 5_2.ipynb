```python
# Напишите функцию area_and_length(), которая принимает на вход строку - радиус окружности,
# а возвращает площадь круга и длину окружности, высчитанных по заданному радиусу,

import math
def area_and_length(stdin):
    stdin = float(stdin)
    circle_area = round(math.pi * stdin**2, 2)
    circle_length = round(2 * math.pi * stdin, 2)
    return f'Circle area = {circle_area}, circle length = {circle_length}'
```
