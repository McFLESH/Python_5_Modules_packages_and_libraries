```python
# Напишите функцию random_integer(), которая принимает на вход 2 числа,
# а возвращает случайное целое число из диапазона чисел, который формируется из поданных на вход двух чисел.
import random
def random_integer(stdin1, stdin2):
    return random.randint(stdin1, stdin2)
```
