```python
# Напишите функцию-таймер, которая принимает на вход строку - время в секундах (например, s),
# а возвращает следующую строку по истечении этого времени:
import time
def timer(stdin):
    stdin = int(stdin)
    time.sleep(stdin)
    return f'Время истекло! {stdin} секунд прошло.'
```
