```python
# Напишите функцию discriminant_root(), которая принимает на вход строку с заданными целыми коэффициентами a, b, c,
# вычисляет корень дискриминанта в уравнении и возвращает это число.
import math
def discriminant_root(k):
    a, b, c = map(int, k.split())
    D = b ** 2 - 4 * a * c
    if D < 0:
        return "Дискриминант меньше нуля"
    return D ** 0.5
```
